```python
class Car():
  def __init__(self,name,color):
    self.name = name                #represent the instance of class named Car
    self.color = color

  def description(self):
    return self.name, self.color
  
  def display(self):
    print("The name and color of the car",self.description())

obj1 = Car("Nissan","white",)       #to create an object with it's atrribute values
obj1.display()
      
```

Application 1 - Write a python program to compute the area and perimeter of a rectangle. Use rectangle as class name, and length and width as attributes


```python
class rectangle():
  def __init__(self,length,width,area):
    self.length = length
    self.width = width

  def area(self):
   return self.length*self.width

  def perimeter(self):
    return 2*(self.length*self.width)
  
  def display(self):
    print("The are of rectangle is",self.area())
    print("The perimeter of rectangle is", self.perimeter())

pol = rectangle(7,2,4)
pol.display()
```
