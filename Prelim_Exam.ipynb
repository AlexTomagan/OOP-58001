```python
class studentwithattributes:
  pass 
  def __init__(self,fullname,studentnumber,age,course):
    self.fullname = fullname
    self.studentnumber = studentnumber
    self.age = age
    self.course = course

  def description(self):
    return self.fullname,self.studentnumber,self.age,self.course

  def display(self):
    print("Student with attributes",self.description())

obj1 = studentwithattributes("Tomagan Alexander","202118463","20","BS CpE")
obj1.display()
```
