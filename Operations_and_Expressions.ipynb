```python
if v<z:
  print("v is less than z")
elif v>z:
  print("v is greater than z")
```

```python
u = 41
if u>10:
  print("u is above 10")
  if u>20:
    print("u is above 20")
else:
  print("u is below 10")
  print("u is below 20")
     
```

Loop Structure

```python
week = ['sunday','saturday','monday','tuesday']
season = ["cloudy, sunny"]

for x in week:
  for y in season:
    print(x,y)
```

```python
for x in week:
  if x =="Thursday":
    break
  print(x)
```
